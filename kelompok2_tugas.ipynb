## sinyal 

Apa Itu Sinyal?
Sinyal adalah fungsi yang membawa informasi. Dalam konteks teknik elektro, sinyal biasanya berupa fungsi waktu yang menunjukkan perubahan besaran fisik seperti tegangan, arus, suara, atau cahaya.
Intinya, sinyal adalah "pembawa informasi" yang bisa diukur dan dianalisis. 
Sinyal bisa berbentuk:
- Sinyal analog: kontinu dalam waktu dan amplitudo (contoh: suara manusia, gelombang radio).
- Sinyal digital: diskrit dalam waktu dan nilai (contoh: sinyal hasil sampling dari mikrofon atau kamera). 

# Contoh Sinyal di Kehidupan Sehari-hari:
- Sinyal suara  :Suara manusia yang terekam oleh mikrofon
- Sinyal listrik  :Tegangan yang dikirim dalam kabel listrik
- Sinyal digital  :Data biner (0 dan 1) dalam komputer atau internet
- Sinyal gambar/video  :Piksel dalam citra atau video yang berubah terhadap waktu
- Sinyal biomedis  :Detak jantung (EKG), gelombang otak (EEG)


# Representasi Sinyal
1. Domain Waktu (Time Domain)
   - Menunjukkan bagaimana nilai sinyal berubah terhadap waktu.
   - Contoh: grafik gelombang suara.

2. Domain Frekuensi (Frequency Domain)
    - Menunjukkan komponen frekuensi yang membentuk sinyal.
    - Contoh: spektrum suara (nada rendah, nada tinggi).

# Instantiate Cosine and Sine Signals
Kalimat ini berasal dari dunia pemrosesan sinyal digital atau pemrograman sinyal.
- Instantiate artinya membuat / membentuk suatu objek berdasarkan rumus atau fungsi tertentu.
- Cosine and sine signals artinya sinyal berbentuk gelombang cosinus dan sinus.

Jadi, "Instantiate cosine and sine signals" berarti:
Membuat sinyal gelombang cosinus dan sinus berdasarkan fungsi matematisnya.

# Apa Itu Sinyal Sinus dan Cosinus?
Sinyal sinus dan cosinus adalah bentuk gelombang dasar dalam matematika dan teknik, sering digunakan untuk:
- Representasi suara,
- Gelombang radio,
- Analisis frekuensi,
- Dan sistem dinamis lainnya.

```python
def plot_signal(signal, sr=1000, t=1, samples_to_show=500, title="Signal", xlabel="Time (s)", ylabel="Amplitude"):
    plt.figure(figsize=(10, 4))
    time_axis = np.linspace(0, t, len(signal))  # ensure time axis matches signal length
    plt.plot(time_axis[:samples_to_show], signal[:samples_to_show])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
duration = 1
sr = 500
freq = 5

t = np.linspace(0, duration, int(sr * duration), endpoint=False)
sine_signal = np.cos(2 * np.pi * freq * t)

plot_signal(sine_signal, sr, t=duration, samples_to_show=2000)
```

```python
duration = 1
sr = 500
freq = 5

t = np.linspace(0, duration, int(sr * duration), endpoint=False)
A = 1 # amplitude
sine_signal = A * np.sin(2 * np.pi * freq * t)

plot_signal(sine_signal, sr, t=duration, samples_to_show=2000)
```

## Mixing Signal ‚Äî Pencampuran Sinyal
Apa Itu "Mixing Signal"?
Mixing signal adalah proses menggabungkan dua atau lebih sinyal untuk menghasilkan sinyal baru.
Tujuan dari mixing bisa bermacam-macam, tergantung aplikasinya: untuk komunikasi, audio, musik, atau analisis frekuensi.

# Jenis Mixing dalam Pemrosesan Sinyal
1. Additive Mixing (Penjumlahan)
Sinyal-sinyal dijumlahkan secara langsung.

Digunakan dalam audio, perekaman suara, dan efek musik.
ùë•(ùë°)=ùë•1(ùë°)+ùë•2(ùë°)x(t)=x1‚Äã(t)+x2‚Äã(t)

2. Multiplicative Mixing (Perkalian / Modulasi)
Sinyal dikalikan satu sama lain.

Digunakan dalam komunikasi radio (modulasi AM, FM, dsb).
ùë•(ùë°)=ùë•1(ùë°)‚ãÖùë•2(ùë°)x(t)=x 1‚Äã(t)‚ãÖx 2‚Äã(t)

Contoh: mengalikan sinyal suara dengan gelombang pembawa (carrier wave) agar bisa dipancarkan lewat antena.

# Aplikasi Mixing Sinyal
- Audio  : Menggabungkan suara vokal dan musik latar dalam rekaman
- Radio  : Modulasi sinyal agar bisa dikirim melalui frekuensi tertentu
- Radar/Sensor  : Menggabungkan sinyal pantulan untuk mendeteksi posisi objek
- DSP (Digital Signal Processing)  : Analisis spektrum, penggabungan channel, dan lain-lain

Jadi, Mixing signal adalah proses mencampurkan dua atau lebih sinyal untuk membentuk sinyal baru. Bisa melalui penjumlahan (additive) atau perkalian (multiplicative/modulasi), tergantung tujuan penggunaannya.

```python
sine_wave = np.sin(2 * np.pi * freq * t)
cosine_wave = np.cos(2 * np.pi * freq * t)

mixed_signal = 0.5 * sine_wave + 0.5 * cosine_wave
rectified_signal = np.abs(mixed_signal)

plot_signal(rectified_signal, sr, t=duration, samples_to_show=1000)
```

```python
sr = 1000          # sample rate (Hz)
duration = 1.0     # seconds
freq = 5           # base frequency (Hz)
t = np.linspace(0, duration, int(sr * duration), endpoint=False)

# Build a sharp waveform by adding harmonics
# Fundamental + 2nd + 3rd harmonic
wave = (
    1.0 * np.sin(2 * np.pi * freq * t) +
    0.5 * np.sin(2 * np.pi * 2 * freq * t) +
    0.33 * np.sin(2 * np.pi * 3 * freq * t)
)

# Normalize (optional)
wave /= np.max(np.abs(wave))
plot_signal(wave, sr, t=duration, samples_to_show=1000)
```

## Waves
Apa Itu Wave (Gelombang) dalam Pemrosesan Sinyal?
Wave atau gelombang adalah bentuk nyata (hasil pengambilan data) dari sinyal yang sudah diambil pada titik-titik waktu tertentu secara berurutan.
Kalau sinyal itu seperti fungsi matematika yang bisa digambarkan sepanjang waktu, maka wave adalah hasil pengambilan data dari sinyal tersebut pada interval waktu yang tetap.

# Contoh Simpelnya
Bayangkan kamu punya sinyal suara yang terus berlangsung.
- Kalau kamu mendengarkannya langsung, itu adalah sinyal analog yang kontinyu.
- Tapi kalau kamu merekamnya (misalnya dengan mikrofon), maka alat perekam akan mengambil cuplikan (sample) dari suara itu setiap detik.
Hasil dari cuplikan-cuplikan itu disebut wave ‚Äî sebuah deretan nilai yang mewakili bagaimana sinyal berubah seiring waktu.

# Karakteristik Wave
- Terbentuk dari banyak sampel: Setiap nilai dalam wave adalah "cuplikan" dari sinyal pada waktu tertentu.
- Jarak antar sampel tetap: Pengambilan nilai dilakukan secara rutin, misalnya setiap 1/44100 detik.
- Bisa disimpan dan diputar ulang: Misalnya dalam file audio seperti WAV atau MP3.

# Apa Itu framerate?
Framerate atau sampling rate adalah berapa banyak sampel diambil setiap detik.
Contohnya: 44.100 sampel per detik (44.1 kHz) adalah -standar kualitas CD audio.

Semakin tinggi framerate-nya:
- Semakin detail gelombangnya,
- Semakin mirip dengan sinyal aslinya,
- Tapi juga membutuhkan lebih banyak data untuk disimpan.

# Contoh dalam Kehidupan Sehari-hari
Audio digital        : Lagu MP3, rekaman suara
Video/audio editing  : Tampilan gelombang suara di aplikasi seperti Audacity
Ilmu fisika          : Gelombang cahaya, gelombang suara

Jadi, Wave adalah representasi nyata dari sinyal yang sudah diambil secara berkala.
Setiap titik pada wave mewakili nilai sinyal pada saat tertentu, dan framerate menentukan seberapa sering pengambilan itu dilakukan.


```python
def signal_function(t, freq=5):
    """Ideal continuous signal: sine wave with frequency `freq` Hz."""
    return np.sin(2 * np.pi * freq * t)

t_cont = np.linspace(0, 1, 1000) 
signal_values = signal_function(t_cont)

sr = 50  
t_wave = np.linspace(0, 1, sr, endpoint=False)
wave_values = signal_function(t_wave)

plt.figure(figsize=(10, 5))
plt.plot(t_cont, signal_values, label='Signal (continuous)', color='blue')
plt.stem(t_wave, wave_values, linefmt='C1-', markerfmt='C1o', basefmt=' ', label='Wave (sampled)')
plt.title("Signal vs Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Apa Itu Spektrum?

Spektrum adalah representasi dari sinyal dalam domain frekuensi. Dengan kata lain, spektrum menunjukkan seberapa banyak energi (atau kekuatan) yang dimiliki sinyal pada masing-masing frekuensi.

# Analisis spektrum penting dalam banyak bidang, seperti:
- Teknologi komunikasi        : untuk memahami sinyal radio atau internet.
- Pengolahan audio            : seperti efek suara, penghilangan noise.
- Pengolahan citra dan video  : untuk kompresi dan pemfilteran.
- ilmu fisik                  : untuk memahami sifat-sifat partikel subatom

# Jenis-Jenis Spektrum
-Spektrum Magnitudo              : menunjukkan seberapa kuat tiap frekuensi (tanpa memperhatikan fase).\
-Spektrum Fase                   : menunjukkan pergeseran fasa dari tiap komponen frekuensi.
-Spektrum Tenaga (Power Spectrum): menunjukkan energi sinyal pada tiap frekuensi (digunakan dalam banyak aplikasi komunikasi dan radar).

# Mengapa Spektrum Penting?
Spektrum sangat penting karena:
- Membantu analisis sistem (misalnya filter, komunikasi).
- Digunakan dalam pengolahan sinyal audio dan gambar.
- Membantu dalam pemfilteran sinyal (menghilangkan noise).
- Mendeteksi sinyal tersembunyi dalam noise (spektrum bisa mengungkap pola tak terlihat di domain waktu).

# Contoh aplikasi nyata:
- Diagnosa suara mesin rusak dari spektrum audio.
- Menganalisis kualitas jaringan radio atau sinyal Wi-Fi.
- Membuat equalizer audio yang mengatur level frekuensi suara.

```python
N = len(wave_values)
fft_vals = np.fft.rfft(wave_values)
frequencies = np.fft.rfftfreq(N, d=1/sr)
magnitude = np.abs(fft_vals)

# 6. Plot Spectrum
plt.figure(figsize=(10, 4))
plt.stem(frequencies, magnitude, basefmt=' ')
plt.title("Spectrum of Sampled Wave")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Time-Domain Signal (Sinyal Domain Waktu)
Time Domain adalah cara merepresentasikan sinyal berdasarkan perubahan amplitudo terhadap waktu. Artinya, kita melihat bagaimana bentuk sinyal berubah seiring waktu. Time domain cocok untuk melihat perubahan sinyal secara langsung.
Misalnya sinyal audio (suara manusia) direkam, lalu diplot dengan sumbu X sebagai waktu (detik) dan sumbu Y sebagai amplitudo (besar suara).

Ciri-ciri:
- Memperlihatkan bentuk gelombang asli.
- Berguna untuk analisis visual dan pemrosesan langsung seperti pemangkasan, filter waktu, dll.
- Tidak menunjukkan informasi frekuensi (berapa cepat sinyal berosilasi).

# Frequency-domain signal (Sinyal Domain Frekuensi)
Frequency domain adalah representasi sinyal berdasarkan komponen frekuensi yang menyusunnya, bukan terhadap waktu. Biasanya diperoleh dengan Transformasi Fourier dari sinyal time domain. Frequency domain lebih cocok untuk melihat komposisi spektral dan melakukan analisis lanjutan. Misalnya, Sinyal audio bisa memiliki berbagai frekuensi: bass (rendah), mid, dan treble (tinggi). Dalam domain frekuensi, kita bisa melihat frekuensi mana yang dominan dalam sinyal.  

Ciri-ciri:
- Menampilkan berapa banyak energi atau kekuatan yang dimiliki oleh setiap frekuensi.
- Cocok untuk analisis seperti filter frekuensi, kompresi sinyal, identifikasi pola, dll.
- Lebih abstrak, tapi sangat kuat untuk pemrosesan sinyal modern (audio, gambar, radar, dll).


```python
sr = 1000          # Sample rate in Hz
duration = 1.0     # Duration in seconds
freq = 5           # Frequency of sine wave (Hz)

# Time axis
t = np.linspace(0, duration, int(sr * duration), endpoint=False)

# Generate sine wave (time-domain)
sine_wave = np.sin(2 * np.pi * freq * t)

# Compute frequency-domain using FFT
N = len(sine_wave)
fft_vals = np.fft.rfft(sine_wave)
frequencies = np.fft.rfftfreq(N, d=1/sr)
magnitude = np.abs(fft_vals)

# Plot Time-Domain Signal
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(t[:200], sine_wave[:200])  # show 200 samples for clarity
plt.title("Time-Domain Signal (Sine Wave)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

# Plot Frequency-Domain Signal
plt.subplot(1, 2, 2)
plt.stem(frequencies, magnitude, basefmt=" ")
plt.title("Frequency-Domain Signal (FFT of Sine Wave)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Fourier Transform

Fourier transform adalah cara untuk mengurai atau membongkar sebuah sinyal menjadi bagian-bagian frekuensinya. Bayangkan ketika kamu mendengar musik, musik itu sebenarnya gabungan dari banyak suara/nada yang dimainkan pada waktu bersamaan. Jadi, Fourier Transform ini seperti alat untuk pemisah nada

Contoh nyatanya:

1.Suara atau musik
Kalau kita merekam  suara, maka hasilnya berupa sinyal waktu (grafik naik turun). Fourier Transform bisa  cari tahu frekuensi apa saja yang muncul di suara itu cocok untuk tuning alat musik, mengenali suara, atau mendeteksi gangguan

2.Foto atau gambar
Di pengolahan gambar, seperti JPEG, gambar pecah jadi frekuensi supaya bisa dikompres, itu juga menggunakan versi dua dimensi dari Fourier Transform

3.Kesehatan
Alat MRI di rumah sakit juga menggunakan Fourier Transform untuk mengubah sinyal magnetik jadi gambar tubuh manusia

Fourier Transform juga memiliki versi cepatnya yang disebut FFT(Fast Fourier Transform), dan versi untuk data digital yaitu DFT (Discrete Fourier Transform). Setelah sebuah sinyal diubah ke bentuk frekuensi, kadang kita ingin memodifikasinya, lalu menngembalikannya ke bentuk semula. untuk itu, kita menggunakan proses kebalikannya yang disebut Inverse Fourier Transform. Jadi, Fourier Transform bukan hanya alat analisis, tapi juga alat rekonstruksi sinyal.



```python
def analyze_wave(sr, duration=1.0, freq=5):
    t = np.linspace(0, duration, int(sr * duration), endpoint=False)
    wave = signal_function(t, freq)

    # FFT
    N = len(wave)
    fft_vals = np.fft.rfft(wave)
    freqs = np.fft.rfftfreq(N, d=1/sr)

    magnitude = np.abs(fft_vals)
    power = magnitude ** 2
    phase = np.angle(fft_vals)

    # Plot time domain
    plt.figure(figsize=(10, 3))
    plt.plot(t, wave)
    plt.title(f"Waveform (Sampling Rate = {sr} Hz)")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plot magnitude spectrum
    plt.figure(figsize=(10, 3))
    plt.stem(freqs, magnitude, basefmt=' ')
    plt.title("Magnitude Spectrum")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plot power spectrum
    plt.figure(figsize=(10, 3))
    plt.stem(freqs, power, basefmt=' ')
    plt.title("Power Spectrum")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Power")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plot phase spectrum
    plt.figure(figsize=(10, 3))
    plt.stem(freqs, phase, basefmt=' ')
    plt.title("Phase Spectrum")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Phase (radians)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
for sample_rate in [20, 50, 100, 1000]:
    analyze_wave(sr=sample_rate, freq=5)
```

```python
from scipy.signal import butter, filtfilt
import IPython.display as ipd

def butter_filter(data, sr, cutoff, btype='low', order=5):
    nyq = 0.5 * sr
    normal_cutoff = np.array(cutoff) / nyq if isinstance(cutoff, (list, tuple)) else cutoff / nyq
    b, a = butter(order, normal_cutoff, btype=btype, analog=False)
    return filtfilt(b, a, data)
```

```python
file_path = "data/PIANO.wav"
y, sr = librosa.load(file_path, sr=None)

duration_sec = len(y) / sr
print(f"Duration: {duration_sec:.2f} seconds")

# Use 0.5s from a valid time (e.g., start at 0.5s)
start_time = 1
end_time = start_time + 1

# Convert to sample index and clip within bounds
start_sample = int(start_time * sr)
end_sample = int(min(len(y), end_time * sr))

segment = y[start_sample:end_sample]

# Check if segment has sufficient data
if len(segment) == 0:
    raise ValueError("Segment is empty. Choose a valid start time and ensure the signal is long enough.")

# Plot spectrum
N = len(segment)
fft_vals = np.fft.rfft(segment)
frequencies = np.fft.rfftfreq(N, d=1/sr)
magnitude = np.abs(fft_vals)

plt.figure(figsize=(10, 4))
plt.plot(frequencies, magnitude)
plt.title("Original Spectrum (0.5s segment)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.tight_layout()
plt.show()

# Listen to original segment
ipd.Audio(segment, rate=sr)
```

# Low-pass filter
Filter ini melewatkan sinyal dengan frekuensi di bawah 1000 Hz dan meredam frekuensi di atas 1000 Hz. Filter ini berguna untuk menghilangkan noise frekuensi tinggi (seperti suara desis atau dengungan) dan juga memfokuskan pada informasi penting di frekuensi rendah, misalnya suara vokal dasar. Dampaknya, hasil suara terdengar lebih "dalam" atau "bas", karena frekuensi tinggi diredam.

# High-pass filter 
Filter ini melewatkan sinyal dengan frekuensi di atas 1000 Hz dan meredam frekuensi di bawah 1000 Hz. Fungsi dari filter ini adalah untuk menghilangkan hum atau noise di frekuensi rendah sehingga memperjelas bagian-bagian sinyal dengan frekuensi tinggi seperti konsonan dalam ucapan atau bunyi instrumen bernada tinggi

# Band-stop filter
Filter ini menghilangkan sinyal dalam rentang frekuensi 500‚Äì2000 Hz dan melewatkan frekuensi di luar rentang tersebut.
Fungsi ini berguna untuk menghapus frekuensi spesifik yang dianggap mengganggu, seperti dengungan atau resonansi di tengah spektrum. Fungsi ini juga digunakan saat ingin mengisolasi komponen frekuensi rendah dan tinggi, sambil menghilangkan bagian tengah.

```python
from scipy.io.wavfile import write
# Low-pass filter (below 1000 Hz)
low = butter_filter(segment, sr, cutoff=1000, btype='low')
ipd.display(ipd.Audio(low, rate=sr))
write("low_pass_output.wav", sr, low)  # Simpan sebagai file WAV

# High-pass filter (above 1000 Hz)
high = butter_filter(segment, sr, cutoff=1000, btype='high')
ipd.display(ipd.Audio(high, rate=sr))
write("high_pass_output.wav", sr, high)  # Simpan sebagai file WAV

# Band-stop filter (remove 500‚Äì2000 Hz)
bandstop = butter_filter(segment, sr, cutoff=[500, 2000], btype='bandstop')
ipd.display(ipd.Audio(bandstop, rate=sr))
write("bandstop_output.wav", sr, bandstop)  # Simpan sebagai file WAV
```

Timbre (warna suara) berubah karena filter memengaruhi komponen frekuensi dari sinyal. Setiap filter menghilangkan atau mengurangi sebagian spektrum frekuensi, yang menyebabkan suara terdengar berbeda:
‚Ä¢	Low-pass filter: Suara terdengar lebih lembut atau gelap (muffled) karena hanya frekuensi rendah (<1000 Hz) yang dipertahankan. Harmonik tinggi dihilangkan.
‚Ä¢	High-pass filter: Suara terdengar lebih tajam atau terang (bright/harsh) karena hanya frekuensi tinggi (>1000 Hz) yang lolos. Nada dasar dan harmonik rendah dihilangkan.
‚Ä¢	Band-stop filter: Suara kehilangan sebagian "isi" atau kedalaman di tengah frekuensi karena frekuensi antara 500‚Äì2000 Hz dibuang. Ini bisa membuat suara terdengar kosong atau aneh, tergantung konten audio.

# Contoh masing-masing filter pada citra 2D

## Low-pass filter

```python
import cv2
from PIL import Image
import requests
from io import BytesIO

# Download gambar Cameraman
img_pil = Image.open('data/Gambar1.jpg').convert("L")  # Grayscale
img = np.array(img_pil)
img = cv2.resize(img, (256, 256))

# FFT
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

# Masking
rows, cols = img.shape
crow, ccol = rows // 2 , cols // 2
radius = 30

low_pass_mask = np.zeros((rows, cols), np.uint8)
cv2.circle(low_pass_mask, (ccol, crow), radius, 1, thickness=-1)
high_pass_mask = 1 - low_pass_mask

# Apply filter
f_low = fshift * low_pass_mask
f_high = fshift * high_pass_mask

# Inverse FFT
img_low = np.abs(np.fft.ifft2(np.fft.ifftshift(f_low)))
img_high = np.abs(np.fft.ifft2(np.fft.ifftshift(f_high)))

# Tampilkan hasil
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(img, cmap='gray')
axs[0].set_title("Original")
axs[1].imshow(img_low, cmap='gray')
axs[1].set_title("Low-pass Filter")
axs[2].imshow(img_high, cmap='gray')
axs[2].set_title("High-pass Filter")
for ax in axs:
    ax.axis('off')
plt.tight_layout()
plt.show()
```

## High-pass filter

```python
blur = cv2.GaussianBlur(img, (11,11), 0)
high_pass = cv2.subtract(img, blur)

plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(1, 2, 2), plt.imshow(high_pass, cmap='gray'), plt.title('High-pass Filtered')
plt.show()
```

## Band-pass filter

```python
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

# Buat band-stop filter di domain frekuensi
rows, cols = img.shape
crow, ccol = rows//2 , cols//2
mask = np.ones((rows, cols), np.uint8)
r_out, r_in = 60, 20  # Rentang radius bandstop
for i in range(rows):
    for j in range(cols):
        d = np.sqrt((i-crow)**2 + (j-ccol)**2)
        if r_in < d < r_out:
            mask[i,j] = 0

fshift_filtered = fshift * mask
img_back = np.fft.ifft2(np.fft.ifftshift(fshift_filtered))
img_back = np.abs(img_back)

plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(1, 2, 2), plt.imshow(img_back, cmap='gray'), plt.title('Band-stop Filtered')
plt.show()
```
